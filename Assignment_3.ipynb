Q2. Sales Data Analysis

Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])

Tasks:
1.
Compute total sales.

2.
Compute mean and standard deviation.

3.
Identify days where sales exceeded the average.

4.
Normalize the sales using Min-Max scaling.

```python
#importing libraries
import numpy as np
import pandas as pd
sales = np.array([120,150,180,200,170,160,210,220,190,175])
```

```python
# total sales.
total_sales=np.sum(sales)
print(total_sales)
```

```python
#mean and standard deviation
sales_mean=np.mean(sales)
sales_deviation=np.std(sales)
print(sales_mean)
print(sales_deviation)
```

```python
# days where sales exceeded the average
average_sales=np.average(sales)
print(average_sales)
for i in sales:
  if (i > average_sales):
    print(i)
```

```python
# min max scaling
min_val=np.min(sales)
max_val=np.max(sales)
scaled_data=(sales-min_val)/(max_val-min_val)
print(scaled_data)
```

Q5. Missing Data Handling

Dataset:

data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}

Tasks:

1.
Detect missing values.

2.
Replace missing Age with mean age.

3.
Replace missing Salary with median salary

4. Remove rows with any remaining missing valuesEmployee

```python
import pandas as pd
data = { 'Name':['John','Anna','Peter','Linda','David'], 'Age':[28, np.nan, 35, 32, np.nan], 'Salary':[50000,54000,np.nan,58000,60000] }
emp_df=pd.DataFrame(data)
print(emp_df)
```

```python
# detect missing values
emp_df.isnull()
```

```python
emp_df.isna().sum()
```

```python
# replace missing age  with mean age
emp_df['Age'].fillna(emp_df['Age'].mean(),inplace=True)
emp_df
```

```python
# replace missing Salary with median salary
emp_df['Salary'].fillna(emp_df['Age'].median(),inplace=True)
emp_df
```

```python
#Remove rows with any remaining missing values
emp_df_new=emp_df.dropna(axis=0)
emp_df_new
```

Q7. Customer Purchase Analysis

Dataset:

data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
Tasks:

1.
Find the average purchase amount by city.

2.
Identify customers with purchase above overall average.

3.
Find the city generating the highest revenue.


4.
Create a column Age_Group:
o
<30 → Young
o
30–50 → Adult
o
50 → Senior

5.
Find average purchase per age group.

```python
import pandas as pd
data = { 'CustomerID':[101,102,103,104,105,106,107], 'Age':[25,45,35,50,23,40,60], 'PurchaseAmount':[250,600,400,800,150,700,900], 'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai'] }
cust_df=pd.DataFrame(data)
cust_df
```

```python
#average purchase amount by city
cust_df.groupby('City')['PurchaseAmount'].mean()
```

```python
#customers with purchase above overall average
avg_purchase =cust_df['PurchaseAmount'].mean()
avg_purchase
```

```python
cust_df[cust_df['PurchaseAmount']>avg_purchase]
```

```python
# customers 102,104,107 and 106
```

```python
#city generating the highest revenue
cust_df.groupby('City')['PurchaseAmount'].sum()

# Ans: Chennai  is the city
```

```python
#new column age_group
cust_df
```

```python
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 50:
        return 'Adult'
    else:  # age >= 50
        return 'Senior'

# Create the Age_Group column
cust_df['Age_Group'] = cust_df['Age'].apply(categorize_age)
cust_df
```

```python
# average purchase per age group
cust_df.groupby('Age_Group')['PurchaseAmount'].mean()
```

Q8. Time Series Sales Analysis

Generate data:

dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})

Tasks:

1.
Convert Date to datetime index.

2.
Compute monthly percentage growth.

3.
Calculate 3-month rolling average sales.

4.
Identify the month with highest growth rate.

```python
import pandas as pd
import numpy as np
```

```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
print(df)
```

```python
df.info()
df.dtypes
```

```python
# Convert Date to datetime index
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)
print(df)
```

```python
print(df.dtypes)
```

```python
#monthly percentage growth
df['Monthly_Growth_%']= df['Sales'].pct_change() * 100
df
```

```python
#3-month rolling average sales.
df['3M_Rolling_Avg'] = df['Sales'].rolling(window=3).mean()
print(df)
```

```python
# month with highest growth rate
max_growth_rate= df['Monthly_Growth_%'].max()
max_growth_rate

# Its march with 13.63
```
