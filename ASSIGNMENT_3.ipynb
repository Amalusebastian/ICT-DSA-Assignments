3. Who scores the most on average for math, reading and writing based on

```python
#Gender
Student_Performance.groupby('gender')[['math score','reading score','writing score']].mean()
```

4 .What do you think about the scoring variation for math, reading and writing based on

● Gender

● Test preparation course

```python
# Gender
Student_Performance.groupby('gender')[['math score','reading score','writing score']].var()
```

```python
#Standard deviation
Student_Performance.groupby('gender')[['math score','reading score','writing score']].std()
```

```python
# min
Student_Performance.groupby('gender')[['math score','reading score','writing score']].min()
```

```python
#max
Student_Performance.groupby('gender')[['math score','reading score','writing score']].max()
```
